Por fim os dados serão transformados em um novo arquivo CSV.

## **Extract**
A extração de dados será feita a partir do arquivo usuarios.csv. O arquivo em questão contém as seguinte colunas `Id`,`Nome`, `G}Gênero de Filmes`.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('usuarios.csv')
users_ids = df['Id'].tolist()

print(df)
```

## **Transform**

Agora será feita a importação da biblioteca necessária e será gerado o primeiro gráfico com os dados atuais. Em seguida será feita a inclusão de mais três usuários e será gerado um novo gráfico.

```python
pip install matplotlib
```

```python
genre_counts = df['Genero_filmes'].value_counts()
print(genre_counts)
```

```python
colors = ['lightcoral','lightblue', 'teal', 'orchid']  # Defina cores personalizadas para as fatias
plt.figure(figsize=(4, 4))  # Define o tamanho da figura

# Cria o gráfico de pizza
plt.pie(genre_counts, labels=genre_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)

plt.title('Distribuição de Gêneros de Filmes Escolhidos')  # Título do gráfico
plt.axis('equal')  # Torna o gráfico de pizza circular

plt.show()  # Exibe o gráfico
```

```python
df = pd.read_csv('usuarios.csv')
novos_usuarios = [
    [11, 'Usuario11', 'Thriller'],
    [12, 'Usuario11', 'Ação'],
    [13, 'Usuario11', 'Comédia'],
     # Adicione quantos novos usuários quiser
]

for usuario in novos_usuarios:
    df = df.append(pd.Series(usuario, index=df.columns), ignore_index=True)

print(df)
```

```python
genre_counts = df['Genero_filmes'].value_counts()
print(genre_counts)
```

```python
colors = ['lightblue','wheat', 'teal', 'orchid', 'lightcoral', 'azure']  # Defina cores personalizadas para as fatias
plt.figure(figsize=(4, 4))  # Define o tamanho da figura

# Cria o gráfico de pizza
plt.pie(genre_counts, labels=genre_counts.index, colors=colors, autopct='%1.1f%%', startangle=180)

plt.title('Gêneros de Filmes Escolhidos')  # Título do gráfico
plt.axis('equal')  # Torna o gráfico de pizza circular

plt.show()  # Exibe o gráfico
```

## **L**oad
